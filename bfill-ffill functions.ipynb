# simple imputer

```python
df=pd.read_csv("C:\\Users\\Vaishnavi\\OneDrive\\Desktop\\thinkryt_internship\\supermarket_sales - Sheet1.csv")
```

```python
from sklearn.impute import SimpleImputer
```

```python
import numpy as np
```

```python
nc=['Date']
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df[nc]=imputer.fit_transform(df[nc])
```
